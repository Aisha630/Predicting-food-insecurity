```python
import matplotlib
import seaborn as sns
import plotly
import chart_studio
import chart_studio.plotly as py
import json
import plotly.express as px
import pandas as pd
```

```python

chart_studio.tools.set_credentials_file(username='xx', api_key='xx')

with open("map_final.geojson", "r") as f:
    geojson = json.load(f)
    
df = pd.read_csv("data/pak_ipc.csv", usecols=[ "District", "Area Phase"])
```

```python
df.head(10)
```

```python
grouped = df.groupby(["District"]).mean().rename(columns={"Area Phase": "value"}) 
```

```python
grouped.head(10)
```

```python
grouped = grouped.reset_index()
```

```python
fig = px.choropleth_mapbox(
    grouped,
    geojson=geojson,
    locations="District",  
    featureidkey="properties.name",  
    color="value",
    mapbox_style="carto-positron",
    zoom=4.5,
    center={"lat": 30.3753, "lon": 69.3451},  
    opacity=0.5,
    title="Average IPC of different districts in Pakistan",
    color_continuous_scale="Viridis", 
)

fig.update_layout(margin={"r":0,"t":30,"l":0,"b":0}, coloraxis_colorbar=dict(
        title="Average IPC", 
        # title_side="top",         
        title_font=dict(size=16, family="Arial"),  
        tickfont=dict(size=12, family="Arial"),   
        lenmode="fraction", len=0.8,              
        thickness=20,                              
        # bgcolor="white",                           
        outlinewidth=1,                           
    ))

fig.show()
```

```python
py.plot(fig, filename='pakistan-districts-map', auto_open=True)
```

```python
fig_json = fig.to_json()
fig_size_bytes = len(fig_json.encode('utf-8'))
fig_size_kb = fig_size_bytes / 1024

print(f"Figure size: {fig_size_kb:.2f} KB")
```
